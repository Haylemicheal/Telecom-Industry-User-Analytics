## Dimensionality Reduction

I use PCA for dimesionality reduction and I created 3 principal components.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
```

```python
data = pca.fit_transform(df.loc[:, ~df.columns.isin(['Start', 'End', 'Last Location Name', 'Handset Manufacturer','Handset Type'])])
```

```python
import pandas as pd
pc_df = pd.DataFrame(data=data, columns=['pca1', 'pca2', 'pca3'])
```
