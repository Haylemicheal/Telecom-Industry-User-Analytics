```python
for i in df.columns:
    percent, _ = util.percent_missing(df[i])
    if (percent > 30):
        df = df.drop([i], axis=1)
```

### If the data is not skewed, filling with either mean or median will work well. But if it is indeed skew, then fill with Median. 

```python
du = DataUnderstanding(df)
du.get_skew()
```

```python
for i in df.columns:
    _, count = util.percent_missing(df[i])
    if count > 0:
        if df[i].dtype == 'object':
            df[i] = df[i].fillna(df[i].mode()[0])
        elif df[i].dtype == 'float64' and df[i].skew(axis=0) != 0:
            df[i] = df[i].fillna(df[i].mode()[0])
        elif df[i].dtype == 'float64' and df[i].skew(axis=0) == 0:
            df[i] = df[i].fillna(df[i].mean())
```
