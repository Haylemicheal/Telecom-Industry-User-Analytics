# Session Duration

```python
session_duration = df.groupby('MSISDN/Number')['Dur. (ms)'].agg([sum])
session_duration = session_duration.sort_values('sum', ascending=False)
print(session_duration.head(20))
```

# Total download (DL) and upload (UL) data

```python
session_duration = df.groupby('MSISDN/Number')[['Total UL (Bytes)', 'Total DL (Bytes)']].sum()
session_duration['Total_data'] = session_duration['Total UL (Bytes)'] + session_duration['Total DL (Bytes)']
session_duration
```

# Total data volume (in Bytes) during this session for each application

```python
cols = df.columns.tolist()
i = cols.index('Social Media DL (Bytes)')

for j in range(i, len(cols) - 1, 2):
    title = cols[j].split(" ")[0] + "_data"
    df[title] = df[cols[j]] + df[cols[j + 1]]
df.columns
```
