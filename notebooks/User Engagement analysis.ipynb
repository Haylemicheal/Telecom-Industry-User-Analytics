## Kmeans Clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(init="random",
                n_clusters=3,
                n_init=10,
                max_iter=300,
                random_state=42
   )
```

```python
model = kmeans.fit(normalized)
```

```python
centroids = kmeans.cluster_centers_
centroids
```

### Group users in k engagement clusters based on the engagement metrics

```python
squared_error = []
kmeans_kwargs= {
    'init':"k-means++",
    'n_init': 10,
    'max_iter': 300,
    'random_state':20  
}
```

```python
for c in range(1, 11):
    kmeans = KMeans(n_clusters=c, **kmeans_kwargs)
    kmeans.fit(normalized)
    squared_error.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.plot(range(1,11), squared_error)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("Squared Error")
plt.title("Elbow Method")
plt.show()
```

#### From the diagram we can see the elbow is in the cluster 3. so the model can give a good accuracy in that cluster.
